The main feature of our gripper is the linear actuator. By means of shoelaces we manage to convert the rotational movement of the motor in a linear move, so do not be surprise if you see some of our member without a shoe. The idea behind is that the depending on the direction of rotation, one string get tight while the other gets loose, pulling the gripper arm one way or another.

The gripper is also able to move up and down thank to a pulley system, we would like to thank the unicorns for sharing that wonderfull idea with us.

Therefore the main purpose of our design is to perform a robust grasping without any sensory feedback.  Thus when a lego piece is sufficiently close to the robot, the gripper will open and go down, then the robot will slowly move forward for certain distance dragging all the lego pieces inside the gripper action space, finally the robot will close the gripper catching all the pieces. 

It's sounds very easy right?. Well the grasping was not as effective as it looked like at first. The gripper didn't have as much friction as needed to hold the pieces. To solve the problem, we use a totally state-of-art material made of a combination between cellulose and sodium sulphate, you may know it as "sponge". This addition increased friction while reducing stiffness in the grasp, making possible the grasping of all different shapes. 

As we all know, everything that shines ain't gold. Our design presented several drawbacks which we corrected. The design was not originally mechanically robust, the shoelaces could get stuck in the gearwheels, we learnt that in the wrong way. Thus we constrained the movements of the shoelaces adding some pegs.
